```python
import os
import sys
import json
import datetime
```

```python

class FreshWorks:
    def __init__(self, path = ""):
        if path == None:
            self.path = os.getcwd()
        else:
            self.path = path
        
        self.__ORGFile="FreshWorks.txt"
        
        try:
            self.path = os.path.join(self.path, self.__ORGFile)
            with open(self.path, "a+") as file:
                file.close()
        
        except FileNotFoundError as fnf_error:
            print("Error Occured : ", fnf_error)
            
        except:
            print('Unable to initialize FreshWorks')
    
    def ifhas(path,key):
        try:
            with open(path,'r') as f:
                para=f.read();
                lines=para.split('\n')
                for line in lines:
                    if line != '':
                        mapp = eval(line)
                        if list(mapp.keys())[0] == key:
                            return true
        
                
        except:
            print('An Error Occured: ')
            if f.closed == False:
                print('File is not closed! ')
            
                
    def elapsedTime():
        end = datetime.datetime.now()
        start = datetime.datetime.combine(end.date(), datetime.time())
        elapsed = (end - start).seconds
        return elapsed
    
    def create(self,key,value,TimeToLive = 86400 ):
        
        if FreshWorks.ifhas(self.path,key) == True:
            print("Key already present!")
        else:
            stats = os.stat(self.path)
            fileSize = (stats.st_size)
            if fileSize >= (1024*1024*1024):
                print("Too Large File!")
                return "Unsuccessful Create!"
            
            dict = {} 
            if type(key) == type("") and len(key) <= 32:
                
                if key.isalpha()==False:
                    print("Key should be alpha numeric!")
                else:
                    try:
                        value = json.loads(value)
                    except ValueError as e:
                        print("Not Valid json object!")
                        return
                    if(sys.getsizeof(value) > (1024*16)):
                        print("json object too large!")
                    else:
                        dict[key] = value
                        tSec = FreshWorks.elapsedTime()
                        if TimeToLive == 86400:
                            dict['TimeToLive'] = TimeToLive - 1
                        else:
                            dict['TimeToLive'] = (tSec+TimeToLive)
                        try:
                            with open(self.path, "a+") as f:
                                f.write(str(dict)+"\n")
                                f.close()
                                print("The value updated in FreshWorks")
                        except:
                            print("Unable to perform create!")
            else:
                print("Key Size too large!")
                
    def read(self,key):
        try:
            with open(self.path,'rt') as f:
                #para = f.read()
                lines = f.read().split('\n')
                for line in lines:
                    if line != '':
                        mapp = eval(line)
                        if list(mapp.keys())[0] == key:
                            if mapp['TimeToLive'] < FreshWorks.elapsedTime():
                                return "json object decayed!"
                            else:
                                return mapp[key]
                #f.close()
        
        except:
            print('An Error Occured: ')
            if f.closed == False:
                print('File is not closed! ')
            
        return "Key Not Found"
                
    def delete(self,key):
        try:
            flag=False
            with open(self.path,'r') as readf:
                lines = readf.readlines()
                readf.close
                
            with open(self.path,'w') as writef:
                for line in lines:
                    mapp=eval(line.strip("\n"))
                    if list(mapp.keys())[0] == key:
                        if mapp['TimeToLive'] < FreshWorks.elapsedTime():
                            writef.write(line)
                        else:
                            flag=True
                            newMap=mapp
                    else:
                        writef.write(line)
            
            if flag:
                print("Delete Successful!")
                return newMap
                
            else:
                print("Delete Unsuccessful!")
                
        
        except:
            print("Something unexpected occured!")
                
                            
                            
                            
```
